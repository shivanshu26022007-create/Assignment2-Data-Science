1. Write a user-defined function that takes a number as input and returns:
<br> 
• "Positive Even"<br> 
• "Positive Odd" <br>
• "Negative" <br>
• "Zero" <br>
Use proper conditional statements and explain your logic.

```python
def number(num):
    if num==0:
        return "Zero"
    elif num<0:
        return "Negative"
    elif num>0 and num % 2==0:
        return "Positive Even"
    else:
        return "Positive Odd"
print(number(16))
print(number(13))
print(number(-35))
print(number(0))
```

3. Write a function that accepts a student’s marks and returns the grade based on the following logic: <br>
• 90+ → A <br>
• 75–89 → B <br>
• 50–74 → C <br>
• Below 50 → Fail <br>
Use nested conditional statements and justify your approach.

```python
def grade(marks):
    if marks>=90 and marks<=100:
        return "A+"
    elif marks>=75 and marks<=89:
        return "B"
    elif marks>=50 and marks<=74:
        return "c"
    elif marks<50 and marks>0:
        return "Fail"
    else:
        return "Plese Enter valid Marks"
print(grade(90))
print(grade(80))
print(grade(72))
print(grade(34))
```

5. Write a function that accepts a list of integers and:<br> 
• Counts even numbers <br>
• Ignores negative numbers<br> 
• Stops processing if zero appears<br> 
Use both loops and conditional statements inside the function. 

```python
def count_even_numbers(lst):
   

    count = 0
    
    for num in lst:          
        if num == 0:         
            break
        if num < 0:          
            continue
        if num % 2 == 0:     
            count += 1
    
    return count

numbers = [4, -2, 7, 8, 0, 10, 12]
result = count_even_numbers(numbers)
print("Count of even numbers:", result)
```
